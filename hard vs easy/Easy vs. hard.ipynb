```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
params = {'legend.fontsize': 'large',
          'figure.figsize': (16, 9),
         'axes.labelsize': 'large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'medium',
         'ytick.labelsize':'medium'}
plt.rcParams.update(params)
import os
from torchvision import transforms
from PIL import Image
```

```python
loader = transforms.Compose([transforms.Resize((300,300))])

images = os.listdir()
filtered = list(filter(lambda x: x.startswith('ILSVRC2012'), images))
types = ['dog', 'flamingo']
level = ['easy', 'hard']

flamingos = list(filter(lambda x: x.startswith('ILSVRC2012_flamingo'), images))
flamingos.sort()

dogs = list(filter(lambda x: x.startswith('ILSVRC2012_dog'), images))
dogs.sort()
```

```python
fig, axes = plt.subplots(2,2, figsize=(9,9))

rows = ['Easy', 'Hard']
for flamingo, ax in zip(flamingos, axes.flat):
    img = Image.open(flamingo)
    img = loader(img)
    ax.imshow(img)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.label_outer()
    
for ax, row in zip(axes[:,0], rows):
    ax.annotate(row, xy=(0, 0.5), xytext=(-ax.yaxis.labelpad - 5, 0),
                xycoords=ax.yaxis.label, textcoords='offset points',
                size='xx-large', ha='right', va='center')
    
fig.tight_layout()
plt.savefig('hard_vs_easy_flamingo.png', dpi=300)
```

```python
fig, axes = plt.subplots(2,2, figsize=(9,9))

rows = ['Easy', 'Hard']
for dog, ax in zip(dogs, axes.flat):
    img = Image.open(dog)
    img = loader(img)
    ax.imshow(img)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.label_outer()
    
for ax, row in zip(axes[:,0], rows):
    ax.annotate(row, xy=(0, 0.5), xytext=(-ax.yaxis.labelpad - 5, 0),
                xycoords=ax.yaxis.label, textcoords='offset points',
                size='xx-large', ha='right', va='center')
    
fig.tight_layout()
plt.savefig('hard_vs_easy_dog.png', dpi=300)
```
