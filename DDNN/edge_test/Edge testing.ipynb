```python
models = ['resnet', 'densenet', 'b-resnet', 'b-densenet', 'msdnet']
```

```python
pc_runs = {
    'b-resnet': pd.read_csv('resnet_localhost.csv', index_col=0)
}
```

```python
cols = pc_runs['b-resnet'].columns.tolist()
cols
```

```python
for data in pc_runs['b-resnet'].groupby(['sample', 'exit']):
    data[1]
```

```python
best_prediction = []
df = pd.DataFrame()
for data in pc_runs['b-resnet'].groupby(['sample']):
    tmp = data[1][np.argmax(np.array(data[1].confidence)):np.argmax(np.array(data[1].confidence))+1].values.tolist()
    df = df.append(dict(zip(cols,tmp[0])), ignore_index=True)
```

```python
best_prediction = []
df = pd.DataFrame()
for data in pc_runs['b-resnet'].groupby(['sample']):
    d = data[1].loc[data[1]['time']<30]
    if not d.empty:
        
        tmp = d[np.argmax(np.array(d.confidence)):np.argmax(np.array(d.confidence))+1].values.tolist()
        df = df.append(dict(zip(cols,tmp[0])), ignore_index=True)
```
